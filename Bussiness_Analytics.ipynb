# Preparing Data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dataset['Ship Mode']= le.fit_transform(dataset['Ship Mode']) 
dataset['Segment']= le.fit_transform(dataset['Segment'])
dataset['Country']= le.fit_transform(dataset['Country'])
dataset['City']= le.fit_transform(dataset['City'])
dataset['State']= le.fit_transform(dataset['State']) 
dataset['Region']= le.fit_transform(dataset['Region']) 
dataset['Category']= le.fit_transform(dataset['Category']) 
dataset['Sub-Category']= le.fit_transform(dataset['Sub-Category']) 
```

```python
x= dataset.iloc[:, :12].values 
```

```python
x
```

```python
y= dataset.iloc[:, 12].values 
```

```python
y
```

# Train and Test Split

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=0)
```

# Craeting and Training the Model

```python
from sklearn.ensemble import RandomForestRegressor 
regressor= RandomForestRegressor(n_estimators= 10, random_state = 0)  
regressor.fit(x_train,y_train)
```

# Prediction and Accuracy Checking

```python
y_pred= regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics 
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) 
print('Root Mean Squared Error:', nm.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
